```python
import time
import urllib
import bs4
import requests
response = requests.get('https://www.letuspublish.com/bollywood-celebrities-and-their-zodiac-sign/')
htmlcontent=response.content
import pandas as pd
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(htmlcontent, 'html.parser')

print(soup.prettify())
#print(soup.get_text)
```

```python
print(soup.find_all('b'))
```

```python
print(soup.find_all('p'))
```

```python
data = [d.find_all('p') for d in soup.find_all('div',{'class':'td-post-content'})]
```

```python
value = []
for i in data:
    for j in i:
        value.append(j.text)

print(value)
```
